```python
import numpy as np

n = int(input("Enter number of seasons: "))

total_matches = 38

multinomial = np.random.multinomial(total_matches, [0.25, 0.25, 0.5], n)

wins = multinomial[:, 0]
draws = multinomial[:, 1]
loses = multinomial[:, 2]

wins_points = wins * 3
draws_points = draws * 1

total_points = wins_points + draws_points

categories = {
    "Below 36": (total_points < 36).sum(),
    "36-62 points": ((total_points >= 36) & (total_points <= 62)).sum(),
    "63-69 points": ((total_points >= 63) & (total_points <= 69)).sum(),
    "70+ points": (total_points >= 70).sum()

}

print("-" * 104)

print(f"Total Points analysis: ")
print(f"Total seasons: {n:_}")
print(f"Max points: {np.max(total_points)}")
print(f"Min points: {np.min(total_points)}")
print(f"Std deviation: {np.std(total_points):.2f}")
print(f"Median points: {np.median(total_points)}")

print("-" * 104)

print("Points Distribution:")
for (category, count) in categories.items():
    probability = count / n
    print(f"{category}: {probability:.2%}  | 1 From {1 / probability:_.1f} seasons")

print("-" * 104)
```
